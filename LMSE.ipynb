## ライブラリのimport

```python
import numpy as np
import matplotlib.pyplot as plt
```

## A. Linear algebra: solve linear equations (M >> N)  
(Slide p.10)

```python
N = 5
M = 50

X = np.random.randn(M, N)
theta = np.random.randn(N, 1)
Y_real = X @ theta
Y_n = Y_real + 4.0 * np.random.randn(M, 1)
theta_hat = np.linalg.inv(X.T @ X) @ X.T @ Y_n
Y_est = X @ theta_hat

plt.plot(Y_real, 'r-', label='Yreal')
plt.plot(Y_n, 'b-', label='Ynoise')
plt.plot(Y_est, 'g-', label='Yestimated')
plt.legend()
plt.show()

plt.plot((Y_n - Y_real) * (Y_n - Y_real), 'r-', label='(Yn-Yreal)^2')
plt.plot((Y_est - Y_real) * (Y_est - Y_real), 'b-', label='(Yest-Yreal)^2')
plt.legend()
plt.show()
```

## B. Gradient descent
(Slide p.11)

```python
N = 5
M = 50

X = np.random.randn(M, N)
theta = np.random.randn(N, 1)
Y_real = X @ theta
Y_n = Y_real + 4.0 * np.random.randn(M, 1)

max_k = 1000
k = 0
alpha = 0.01
theta_hat_k_1 = np.random.randn(N, 1)
theta_hat_k = theta_hat_k_1
xi = 10e-6

while k < max_k:
    e_k = X @ theta_hat_k - Y_n
    if np.linalg.norm(e_k) ** 2 < xi:
        break
    k = k + 1
    theta_hat_k_1 = theta_hat_k
    theta_hat_k = theta_hat_k_1 - alpha * X.T @ e_k
theta_hat = theta_hat_k
Y_est = X @ theta_hat

plt.plot(Y_real, 'r-', label='Yreal')
plt.plot(Y_n, 'b-', label='Ynoise')
plt.plot(Y_est, 'g-', label='Yestimated')
plt.legend()
plt.show()

plt.plot((Y_n - Y_real) * (Y_n - Y_real), 'r-', label='(Yn-Yreal)^2')
plt.plot((Y_est - Y_real) * (Y_est - Y_real), 'b-', label='(Yest-Yreal)^2')
plt.legend()
plt.show()
```
